# Dados Pokedex

## Relacionar dados

```python
df_resumida = df.merge(df_capture_rate, left_on='#', right_on='pokedex_number').copy()
df_resumida.drop(["#", "name", 'Total'], axis=1, inplace=True)
```

## Reordenar colunas

```python
cols = list(df_resumida.columns.values)
cols = [
    'pokedex_number',
    'Name',
    'Type 1',
    'Type 2',
    'HP',
    'Attack',
    'Defense',
    'Sp. Atk',
    'Sp. Def',
    'Speed',
    'capture_rate',
    'Legendary',
    'Generation',
    ]
df_resumida.head(1)
```

```python
df_resumida = df_resumida[cols]
df_resumida.head(1)
```

## Renomear colunas

```python
cols_renamed = {
    'pokedex_number': 'pokedex_number',
    'Name': 'name',
    'Type 1': 'type1',
    'Type 2': 'type2',
    'HP': 'hp',
    'Attack': 'attack',
    'Defense': 'defense',
    'Sp. Atk': 'sp_atk',
    'Sp. Def': 'sp_def',
    'Speed': 'speed',
    'capture_rate': 'capture_rate',
    'Legendary': 'legendary',
    'Generation': 'generation',
}
df_resumida.rename(columns=cols_renamed, inplace=True)
```

## Ataque e Defesa Total

```python
df_resumida["attack_total"] = df_resumida["attack"] + df_resumida["sp_atk"]
df_resumida["defense_total"] = df_resumida["defense"] + df_resumida["sp_def"]

df_resumida.drop(["attack", "sp_atk", "defense", "sp_def"], axis=1, inplace=True)
```

```python
cols = list(df_resumida.columns.values)
cols = [
    'pokedex_number',
    'name',
    'type1',
    'type2',
    'hp',
    'attack_total',
    'defense_total',
    'speed',
    'capture_rate',
    'legendary',
    'generation'
]
df_resumida = df_resumida[cols]
```

## Fill(na)

```python
df_resumida.fillna("NULL", inplace=True)
```

## Definir tipo das colunas

```python
df_resumida["capture_rate"] = df_resumida["capture_rate"].astype('int64')
```

# Matriz Tipo

```python
df = pd.read_excel("data/matriz_tipo.xlsx", sheet_name="tratada")
df
```

```python
covert = {
    "0×": 0,
    "½×": 0.5,
    "1×": 1,
    "2×": 2
}
```

```python
# Replace the values in the DataFrame
for col in df.columns:
    if col != 'tipo':
        df[col] = df[col].map(covert)

df
```

```python
df.to_csv("data/matriz_tipo.csv", index=False)
```
